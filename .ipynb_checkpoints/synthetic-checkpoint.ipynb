```python
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import colorsys
from IPython.display import display
import random
import xml.etree.ElementTree as ET
%matplotlib inline
random.seed(12)
```

```python
def change_hue(im):
    width, height = im.size
    ld = im.load()
    hue_shift = random.choice([15.0,25.0, 90.])
    for y in range(height):
        for x in range(width):
            r,g,b = ld[x,y]
            h,s,v = colorsys.rgb_to_hsv(r/255., g/255., b/255.)
            h = (h + -hue_shift/360.0) % 1.0
            r,g,b = colorsys.hsv_to_rgb(h, s, v)
            ld[x,y] = (int(r * 255.9999), int(g * 255.9999), int(b * 255.9999))
```

```python
def simulate_odo(im):
    new_im = im.copy()
    odo_val = random.randint(2000, 200000)
    trip_odo = random.uniform(1,odo_val/4)
    #odo_val = 11111111
    #trip_odo = 2222
    draw = ImageDraw.Draw(new_im)
    font = ImageFont.truetype("digital-7.ttf", 28)

    # ODO 
    draw.text((350, 330),str(odo_val),(255,255,255),font=font)

    # TRIP ODO
    draw.text((450, 330),"{0:.1f}".format(trip_odo),(255,255,255),font=font)
#     create_annotation(odoval, t_odoval)
    return new_im
```

```python
import cv2
import numpy as np

# Load and display the original image
img = cv2.imread("dashboard2.jpg")
cv2.namedWindow("Original", cv2.WINDOW_NORMAL)
#cv2.imshow("Original", img)
#cv2.waitKey(0)

```

```python
# Create the basic black image 
mask = np.zeros(img.shape, dtype = "uint8")

# Draw a white, filled rectangle on the mask image
cv2.rectangle(mask, (350, 330), (450, 340), (255, 255, 255), -1)

# Display constructed mask
cv2.namedWindow("Mask", cv2.WINDOW_NORMAL)
cv2.imshow("Mask", mask)
cv2.waitKey(0)
```

```python
import numpy as np
import cv2
img = cv2.imread('dashboard2.jpg')
# Create the basic black image 
mask = np.zeros(img.shape, dtype = "uint8")
# Draw a white, filled rectangle on the mask image
mask = cv2.rectangle(mask, (350, 330), (450, 380), (255, 255, 255), -1)
gray_mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
dst = cv2.inpaint(img,gray_mask,3,cv2.INPAINT_TELEA)
cv2.imwrite('dummy.jpg',dst)
```
